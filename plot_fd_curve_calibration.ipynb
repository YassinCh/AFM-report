```python
import pandas as pd 

#Vanwege tijd niet het meest wonderschone code ooit , veel copy paste!.

df_backward= pd.read_csv('data/fd_curve_calibration/fd_curve_middle_ridge_1_backward.csv', sep=';' , header=None, dtype=str)
df_forward = pd.read_csv('data/fd_curve_calibration/fd_curve_middle_ridge_1_forward.csv', sep=';' , header=None, dtype=str)
df_2_backward= pd.read_csv('data/fd_curve_calibration/fd_curve_middle_ridge_2_backward.csv', sep=';' , header=None, dtype=str)
df_2_forward = pd.read_csv('data/fd_curve_calibration/fd_curve_middle_ridge_2_forward.csv', sep=';' , header=None, dtype=str)
df_3_backward = pd.read_csv('data/fd_curve_calibration/fd_curve_middle_ridge_3_backward.csv', sep=';' , header=None, dtype=str)
df_3_forward = pd.read_csv('data/fd_curve_calibration/fd_curve_middle_ridge_3_forward.csv', sep=';' , header=None, dtype=str)

df_backward = df_backward.astype(float)
df_forward = df_forward.astype(float)
df_2_backward = df_2_backward.astype(float)
df_2_forward = df_2_forward.astype(float)
df_3_backward = df_3_backward.astype(float)
df_3_forward = df_3_forward.astype(float)

pd.set_option('display.float_format', '{:.10e}'.format)

df_backward = df_backward.T
df_forward = df_forward.T
df_2_backward = df_2_backward.T
df_2_forward = df_2_forward.T
df_3_backward = df_3_backward.T
df_3_forward = df_3_forward.T
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(10, 6))


x = df_backward[0] * 1e6
y = df_backward[1]

x_2 = df_forward[0] * 1e6
y_2 = df_forward[1]

x_3 = df_2_backward[0] * 1e6
y_3 = df_2_backward[1]

x_4 = df_2_forward[0] * 1e6
y_4 = df_2_forward[1]

x_5 = df_3_forward[0] * 1e6
y_5 = df_3_forward[1]

x_6 = df_3_backward[0] * 1e6
y_6 = df_3_backward[1]


plt.scatter(x, y, color='red', label='backward trace 1',s=1,alpha=0.5)
plt.scatter(x_2, y_2, color='blue', label='forward trace 1',s=1,alpha=0.5)
plt.scatter(x_3, y_3, color='green', label='backward trace 2',s=1,alpha=0.5)
plt.scatter(x_4, y_4, color='yellow', label='forward trace 2',s=1,alpha=0.5)
plt.scatter(x_5, y_5, color='purple', label='forward trace 3', s=1,alpha=0.5)
plt.scatter(x_6, y_6, color='orange', label='backward trace 3', s=1,alpha=0.5)


mask = np.logical_and(x>-3.1 , x<-2.7)
mask_2 = np.logical_and(x_2>-3.1 , x_2<-2.7)
mask_3 = np.logical_and(x_3>-3.2 , x_3<-2.8)
mask_4 = np.logical_and(x_4>-3.2 , x_4<-2.8)
mask_5 = np.logical_and(x_5>-3.3 , x_5<-2.9)
mask_6 = np.logical_and(x_6>-3.3 , x_6<-2.9)

x_segment = x[mask]
y_segment = y[mask]
x_2_segment = x_2[mask_2]
y_2_segment = y_2[mask_2]
x_3_segment = x_3[mask_3]
y_3_segment = y_3[mask_3]
x_4_segment = x_4[mask_4]
y_4_segment = y_4[mask_4]
x_5_segment = x_5[mask_5]
y_5_segment = y_5[mask_5]
x_6_segment = x_6[mask_6]
y_6_segment = y_6[mask_6]

p = np.polyfit(x_segment, y_segment, 1)
p_2 = np.polyfit(x_2_segment, y_2_segment, 1)
p_3 = np.polyfit(x_3_segment, y_3_segment, 1)
p_4 = np.polyfit(x_4_segment, y_4_segment, 1)
p_5 = np.polyfit(x_5_segment, y_5_segment, 1)
p_6 = np.polyfit(x_6_segment, y_6_segment, 1)

y_fit = np.polyval(p, x_segment)
y_fit_2 = np.polyval(p_2, x_2_segment)
y_fit_3 = np.polyval(p_3, x_3_segment)
y_fit_4 = np.polyval(p_4, x_4_segment)
y_fit_5 = np.polyval(p_5, x_5_segment)
y_fit_6 = np.polyval(p_6, x_6_segment)


plt.plot(x_segment, y_fit, color='black', label = 'fits')
plt.plot(x_2_segment, y_fit_2, color='black')
plt.plot(x_3_segment, y_fit_3, color='black')
plt.plot(x_4_segment, y_fit_4, color='black')
plt.plot(x_5_segment, y_fit_5, color='black')
plt.plot(x_6_segment, y_fit_6, color='black')

plt.xlabel('Distance($\mu m$)')
plt.ylabel('Voltage(nV)')
plt.legend()

plt.tick_params(direction='in', length=3, width=1)

plt.grid(True, which='both', linestyle='--', linewidth=0.5, zorder=0)

plt.tight_layout() 

plt.savefig('fd_curve_calibration.pdf',)

plt.show()
```

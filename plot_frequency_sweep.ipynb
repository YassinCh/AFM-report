```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import curve_fit
df_curve_1 = pd.read_csv('data/frequency/frequency_sweep_1.csv', header=None,sep=';')
df_curve_2 = pd.read_csv('data/frequency/frequency_sweep_2.csv', header=None,sep=';')
df_curve_3 = pd.read_csv('data/frequency/frequency_sweep_3.csv', header=None,sep=';')
df_curve_1 = df_curve_1.T
df_curve_2 = df_curve_2.T
df_curve_3 = df_curve_3.T



plt.figure(figsize=(10, 6))
plt.scatter(df_curve_1[0], df_curve_1[1], marker='.', color='b', label='Frequency sweep 1')
plt.scatter(df_curve_2[0], df_curve_2[1], marker='.', color='r', label='Frequency sweep 2')
plt.scatter(df_curve_3[0], df_curve_3[1], marker='.', color='y', label='Frequency sweep 3')

plt.xlabel('Frequency (Hz)')
plt.ylabel('Voltage (V)')

plt.axvline(x=23015, color='g', linestyle='--', label='Resonance frequency (23.02 kHz)')

plt.grid(True)
plt.legend()
plt.savefig('processed_data/frequency_sweep.pdf')
plt.show()

```
